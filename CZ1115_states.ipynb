```python
USACleanData = pd.DataFrame(USAData[['date','death','positive','negative','totalTestResults']])

# Get indexes where totalTestResultsIncrease is still 0 (covid hasn't started)
indexNA = USACleanData[USACleanData['totalTestResults'] == 0].index

# Delete these row indexes from dataFrame
USACleanData.drop(indexNA , inplace=True)
```

```python
latestdata = pd.DataFrame(columns=('date','state', 'positive', 'negative','totalTestResults'))
datanames = ['caCleanData','txCleanData','flCleanData','nyCleanData','ilCleanData','paCleanData','gaCleanData','ohCleanData','njCleanData','ncCleanData','USACleanData']

for data in datanames:
    extractlatest=vars()[data].loc[vars()[data]['date']== '2021-03-07']  #vars() makes the string into a variable name
    latestdata = latestdata.append(extractlatest,ignore_index=True)

#create new row "Others" -> rest of the other states
#subtract data of 10 states from the USA data
USArow = latestdata.iloc[10,2:]
for i in range(10):
    staterows = latestdata.iloc[i,2:]
    USArow = USArow.subtract(staterows,fill_value=0)

#make USArow a dataframe and into a row
USArow = pd.DataFrame(data=USArow).transpose()

#creating 'Others' row
others = {'date':['2021-03-07'],'state':['Others']}
others = pd.DataFrame(data=others)
others = pd.concat([others,USArow],axis=1)

#concatenate others row to the rest of the data (before USA data)
latestdata = pd.concat([latestdata.iloc[:10],others,latestdata.iloc[10:]],ignore_index=True)
latestdata
```

### Required Libaries for Polynomial and SVM Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

### National Cases

```python
USApositive = USA_positive.iloc[:, ::-1]
```

```python
USApositive = pd.DataFrame.transpose(USApositive)
```

```python
USApositive = USApositive.reset_index(drop=True)
```

```python
USApositive = np.array(USApositive)
```

```python
X_train_USApositive, X_test_USApositive, y_train_USApositive, y_test_USApositive = train_test_split(days_0304, USApositive, test_size=0.25, shuffle=False)
```

```python
#to show linear regression is not suitable
lin_reg = LinearRegression()
lin_reg.fit(days_0304, USApositive)


plt.scatter(days_0304, USApositive, color='red')
plt.plot(days_0304, lin_reg.predict(days_0304), color='blue')
plt.title('Positive Cases in California')
plt.xlabel('Days')
plt.ylabel('Positive Cases')
plt.show()
print('MAE:', mean_absolute_error(USApositive, lin_reg.predict(days_0304)))
print('MSE:',mean_squared_error(USApositive, lin_reg.predict(days_0304)))
```

```python
#polynomial regression of degree 5 is most suitable
poly = PolynomialFeatures(degree=5)
poly_days = poly.fit_transform(days_0304)
pol = LinearRegression()
pol.fit(poly_days, USApositive)

plt.scatter(days_0304, USApositive, color='red')
plt.plot(days_0304, pol.predict(poly.fit_transform(days_0304)), color='blue')
plt.title('Positive Cases in USA')
plt.xlabel('Days')
plt.ylabel('Positive Cases')
plt.show()
print('MAE:', mean_absolute_error(USApositive, pol.predict(poly.fit_transform(days_0304))))
print('MSE:',mean_squared_error(USApositive, pol.predict(poly.fit_transform(days_0304))))
```

```python
# transform data for polynomial regression
poly = PolynomialFeatures(degree=5)
poly_X_train_USApositive = poly.fit_transform(X_train_USApositive)
poly_X_test_USApositive = poly.fit_transform(X_test_USApositive)
poly_future_forecast = poly.fit_transform(future_forecast)
```

```python
# polynomial regression
linear_model = LinearRegression(normalize=True, fit_intercept=True)
linear_model.fit(poly_X_train_USApositive, y_train_USApositive)
test_linear_pred = linear_model.predict(poly_X_test_USApositive)
linear_pred = linear_model.predict(poly_future_forecast)
print('MAE:', mean_absolute_error(test_linear_pred, y_test_USApositive))
print('MSE:',mean_squared_error(test_linear_pred, y_test_USApositive))
```

```python
plt.plot(y_test_USApositive)
plt.plot(test_linear_pred)
plt.legend(['Test Data', 'Polynomial Regression Predictions'])
```

```python
#Support Vector Regression

from sklearn.svm import SVR

svm_USApositive = SVR(kernel = 'poly',degree=5, C=0.1)
svm_USApositive.fit(X_train_USApositive,y_train_USApositive)
svm_USApositive_pred = svm_USApositive.predict(future_forecast)
```

```python
svm_test_pred = svm_USApositive.predict(X_test_USApositive)
plt.plot(y_test_USApositive)
plt.plot(svm_test_pred)
plt.legend(['Test Data', 'SVM Predictions'])
print('MAE:', mean_absolute_error(svm_test_pred, y_test_USApositive))
print('MSE:',mean_squared_error(svm_test_pred, y_test_USApositive))
```
