```python
worldCleanData = pd.DataFrame(worldData[['date','death','positive','negative','totalTestResults']])

# Get indexes where totalTestResultsIncrease is still 0 (covid hasn't started)
indexNA = worldCleanData[worldCleanData['totalTestResults'] == 0].index

# Delete these row indexes from dataFrame
worldCleanData.drop(indexNA , inplace=True)
```

```python
worldCleanData
```

```python
latestdata = pd.DataFrame(columns=('date','state', 'positive', 'negative','totalTestResults'))
datanames = ['caCleanData','txCleanData','flCleanData','nyCleanData','ilCleanData','worldCleanData']

for data in datanames:
    extractlatest=vars()[data].loc[vars()[data]['date']== '2021-03-07']  #vars() makes the string into a variable name
    latestdata = latestdata.append(extractlatest,ignore_index=True)

#create new row "Others" -> rest of the other states
#subtract data of 5 states from the world data
worldrow = latestdata.iloc[5,2:]
for i in range(5):
    staterows = latestdata.iloc[i,2:]
    worldrow = worldrow.subtract(staterows,fill_value=0)

#make worldrow a dataframe and into a row
worldrow = pd.DataFrame(data=worldrow).transpose()

#creating 'Others' row
others = {'date':['2021-03-07'],'state':['Others']}
others = pd.DataFrame(data=others)
others = pd.concat([others,worldrow],axis=1)

#concatenate others row to the rest of the data (before World data)
latestdata = pd.concat([latestdata.iloc[:5],others,latestdata.iloc[5:]],ignore_index=True)
latestdata
```
